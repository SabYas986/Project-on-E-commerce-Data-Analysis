### Removing Missing Values & Duplicate Values

```python
conn=databaseConnect()
sql="SELECT * FROM customer"
customerFrame=pd.read_sql(sql,conn)
customerFrame.dropna(inplace=True)
customerFrame.drop_duplicates(inplace=True)

sql="SELECT * FROM product"
productFrame=pd.read_sql(sql,conn)
productFrame.dropna(inplace=True)
productFrame.drop_duplicates(inplace=True)

sql="SELECT * FROM order_details"
orderFrame=pd.read_sql(sql,conn)
orderFrame.dropna(inplace=True)
orderFrame.drop_duplicates(inplace=True)

print("After fixing data....")
print(customerFrame)
print(productFrame)
print(orderFrame)

```

## Exploratory Data Analysis Through Visualization
### Customer Analysis
#### 1.Identify the most frequent customers based on their order history.

```python
customer_count = customerFrame.groupby('city').size().reset_index(name='count')
customer_count = customer_count.sort_values(by='count', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(customer_count['city'], customer_count['count'], color=['skyblue','yellow','red','cyan'])
plt.title('Number of Customers City-wise')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

### Product Analysis
#### 1. Determine the total number of products available by category.

```python
products_count_by_category = productFrame.groupby('category').size().reset_index(name='total_products')
products_count_by_category = products_count_by_category.sort_values(by='total_products', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(products_count_by_category['category'], products_count_by_category['total_products'], color=['skyblue','green','magenta'])
plt.title('Total Number of Products by Category')
plt.xlabel('Category')
plt.ylabel('Total Number of Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 2. Analyze the distribution of products across sub-categories.

```python
products_count_by_subcategory = productFrame.groupby('sub_category').size().reset_index(name='total_products')
products_count_by_subcategory = products_count_by_subcategory.sort_values(by='total_products', ascending=False)
plt.figure(figsize=(12, 8))
plt.bar(products_count_by_subcategory['sub_category'], products_count_by_subcategory['total_products'], color=['red','yellow','skyblue','orange','lightgreen'])
plt.title('Distribution of Products Across Sub-Categories')
plt.xlabel('Sub-Category')
plt.ylabel('Total Number of Products')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
```

### 3. Analyze the distribution of products across sub-categories.

```python
low_stock_threshold = 10
low_stock_products = productFrame[productFrame['stock'] < low_stock_threshold]
low_stock_products = low_stock_products.sort_values(by='stock')
plt.figure(figsize=(12, 8))
plt.bar(low_stock_products['product_name'], low_stock_products['stock'], color=['skyblue','yellow'])
plt.title('Products with Low Stock Levels')
plt.xlabel('Product Name')
plt.ylabel('Stock Level')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

### 4. Identify products with low stock levels.

```python
average_selling_price = productFrame['selling_price'].mean()
max_selling_price = productFrame['selling_price'].max()
min_selling_price = productFrame['selling_price'].min()
prices = [average_selling_price, max_selling_price, min_selling_price]
labels = ['Average Selling Price', 'Maximum Selling Price', 'Minimum Selling Price']
plt.figure(figsize=(8, 8))
plt.pie(prices, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Selling Prices')
plt.axis('equal')
plt.show()
```

### 5. Calculate the average, maximum, and minimum selling prices for products..

```python
merged_data = pd.merge(orderFrame, productFrame, on='product_id')
orders_product_wise = merged_data.groupby('product_name').size().reset_index(name='total_orders')
plt.figure(figsize=(10, 10))
plt.pie(orders_product_wise['total_orders'], labels=orders_product_wise['product_name'], autopct='%1.1f%%', startangle=140)
plt.title('Total Orders Product-wise')
plt.axis('equal')  
plt.show()
```

### 1. Calculate the top 10 orders product wise.

```python
merged_data = pd.merge(orderFrame, productFrame, on='product_id')
orders_product_wise = merged_data.groupby('product_name')['quantity'].sum().reset_index()
top_10_orders = orders_product_wise.nlargest(10, 'quantity')
plt.figure(figsize=(12, 8))
plt.barh(top_10_orders['product_name'], top_10_orders['quantity'], color='skyblue')
plt.title('Top 10 Orders Product-wise')
plt.xlabel('Product Name')
plt.ylabel('Total Quantity Ordered')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

### 2. Analyze the order status distribution (e.g., pending, delivered).

```python
order_status_distribution = orderFrame.groupby('order_status').size().reset_index(name='count')
order_status_distribution = order_status_distribution.sort_values(by='count', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(order_status_distribution['order_status'], order_status_distribution['count'], color=['magenta','cyan','darkgreen'])
plt.title('Order Status Distribution')
plt.xlabel('Order Status')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
```

### 3. Identify the most popular products based on order quantity.

```python
merged_data = pd.merge(orderFrame, productFrame, on='product_id')
product_order_quantities = merged_data.groupby('product_name')['quantity'].sum().reset_index()
most_popular_product = product_order_quantities.nlargest(1, 'quantity')
plt.figure(figsize=(10, 6))
plt.barh(most_popular_product['product_name'], most_popular_product['quantity'], color='skyblue')
plt.title('Most Popular Product based on Order Quantity')
plt.xlabel('Product Name')
plt.ylabel('Total Quantity Ordered')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

### 1. Calculate total revenue generated from orders product wise.

```python
merged_data = pd.merge(orderFrame, productFrame, on='product_id')
merged_data['revenue'] = merged_data['quantity'] * merged_data['selling_price']
product_revenue = merged_data.groupby('product_name')['revenue'].sum().reset_index()
product_revenue = product_revenue.sort_values(by='revenue', ascending=False)
plt.figure(figsize=(12, 8))
plt.bar(product_revenue['product_name'], product_revenue['revenue'], color=['skyblue','orange','lightgreen','red','yellow','magenta'])
plt.title('Total Revenue Generated from Orders Product-wise')
plt.xlabel('Product Name')
plt.ylabel('Total Revenue Generated')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

### 2. Calculate total revenue product category wise percentage.

```python
merged_data = pd.merge(orderFrame, productFrame, on='product_id')
merged_data['revenue'] = merged_data['quantity'] * merged_data['selling_price']
category_revenue = merged_data.groupby('category')['revenue'].sum().reset_index()
total_revenue = category_revenue['revenue'].sum()
category_revenue['percentage'] = (category_revenue['revenue'] / total_revenue) * 100
plt.figure(figsize=(8, 8))
plt.pie(category_revenue['percentage'], labels=category_revenue['category'], autopct='%1.1f%%', startangle=140)
plt.title('Total Revenue Product Category-wise Percentage')
plt.axis('equal')  
plt.show()
```

### 3. Calculate the total revenue generated from all orders.

```python
merged_data = pd.merge(orderFrame, productFrame, on='product_id')
merged_data['revenue'] = merged_data['quantity'] * merged_data['selling_price']
total_revenue = merged_data['revenue'].sum()
plt.figure(figsize=(8, 6))
plt.bar(['Total Revenue'], [total_revenue], color='skyblue')
plt.title('Total Revenue Generated from All Orders')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()
```

### 4. Identify the most profitable products based on the difference between  original and selling prices.

```python
productFrame['profit'] = productFrame['original_price'] - productFrame['selling_price']
most_profitable_products = productFrame.sort_values(by='profit', ascending=False).head(10)  
plt.figure(figsize=(12, 8))
plt.bar(most_profitable_products['product_name'], most_profitable_products['profit'], color='magenta')
plt.title('Top 10 Most Profitable Products')
plt.xlabel('Product Name')
plt.ylabel('Profit (Original Price - Selling Price)')
plt.xticks(rotation=45, ha='right') 
plt.tight_layout()
plt.show()
```

### 1. Identify product names with the highest and lowest order quantities.

```python
merged_data = pd.merge(orderFrame, productFrame, on='product_id')
product_order_quantities = merged_data.groupby('product_name')['quantity'].sum().reset_index()
product_order_quantities_sorted = product_order_quantities.sort_values(by='quantity')
plt.figure(figsize=(12, 8))
plt.bar(product_order_quantities_sorted['product_name'][:10], product_order_quantities_sorted['quantity'][:10], color='skyblue')
plt.bar(product_order_quantities_sorted['product_name'][-10:], product_order_quantities_sorted['quantity'][-10:], color='orange')
plt.title('Top and Bottom Products by Order Quantities')
plt.xlabel('Product Name')
plt.ylabel('Total Quantity Ordered')
plt.xticks(rotation=45, ha='right')  
plt.legend(['Lowest Quantities', 'Highest Quantities'], loc='upper left')
plt.tight_layout()
plt.show()
```

### 2. Identify customers with the highest and lowest order quantities by  customer name.

```python
customer_order_quantities = orderFrame.groupby('customer_id')['quantity'].sum().reset_index()
highest_order_quantities = customer_order_quantities.sort_values(by='quantity', ascending=False).head(10)
lowest_order_quantities = customer_order_quantities.sort_values(by='quantity', ascending=True).head(10)
fig, axes = plt.subplots(2, 1, figsize=(12, 14))
axes[0].bar(highest_order_quantities['customer_id'], highest_order_quantities['quantity'], color='skyblue')
axes[0].set_title('Customers with Highest Order Quantities')
axes[0].set_xlabel('Customer ID')
axes[0].set_ylabel('Total Quantity Ordered')
axes[1].bar(lowest_order_quantities['customer_id'], lowest_order_quantities['quantity'], color='skyblue')
axes[1].set_title('Customers with Lowest Order Quantities')
axes[1].set_xlabel('Customer ID')
axes[1].set_ylabel('Total Quantity Ordered')

plt.tight_layout()
plt.show()
```

### 3. Determine the most preferred payment modes..

```python
payment_mode_distribution = orderFrame.groupby('payment_mode').size().reset_index(name='count')
payment_mode_distribution = payment_mode_distribution.sort_values(by='count', ascending=False)
plt.figure(figsize=(8, 8))
plt.pie(payment_mode_distribution['count'], labels=payment_mode_distribution['payment_mode'], autopct='%1.1f%%', startangle=140)
plt.title('Preferred Payment Modes')
plt.axis('equal')
plt.show()
```

### 1. Identify peak order date.

```python
orderFrame['order_date'] = pd.to_datetime(orderFrame['order_date'])
daily_orders =orderFrame.groupby('order_date').size().reset_index(name='total_orders')
peak_order_date = daily_orders.loc[daily_orders['total_orders'].idxmax(), 'order_date']
plt.figure(figsize=(12, 6))
plt.plot(daily_orders['order_date'], daily_orders['total_orders'], color='skyblue')
plt.axvline(x=peak_order_date, color='red', linestyle='--', label='Peak Order Date: {}'.format(peak_order_date.strftime('%Y-%m-%d')))
plt.title('Total Orders per Day')
plt.xlabel('Date')
plt.ylabel('Total Orders')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 1. Explore the distribution of customers across different cities.

```python
city_distribution = customerFrame['city'].value_counts()
plt.figure(figsize=(12, 6))
city_distribution.plot(kind='bar', color='skyblue')
plt.title('Distribution of Customers Across Cities')
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Identify top 10 slow-moving products based on low sales.

```python
product_sales = orderFrame.groupby('product_id')['total_price'].sum().reset_index()
slow_moving_products = product_sales.sort_values(by='total_price').head(10)
slow_moving_products = pd.merge(slow_moving_products, productFrame, on='product_id')
plt.figure(figsize=(12, 6))
plt.bar(slow_moving_products['product_name'], slow_moving_products['total_price'], color=['skyblue','orange','green','yellow','red','cyan','purple'])
plt.title('Top 10 Slow-Moving Products Based on Low Sales')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Display successful and pending payments order counts.

```python
successful_orders = orderFrame[orderFrame['order_status'] == 'Delivered'] 
pending_orders = orderFrame[orderFrame['order_status'] == 'Pending']
successful_order_counts = successful_orders.groupby('order_date').size().reset_index(name='successful_count')
pending_order_counts = pending_orders.groupby('order_date').size().reset_index(name='pending_count')
plt.figure(figsize=(12, 6))
plt.plot(successful_order_counts['order_date'], successful_order_counts['successful_count'], label='Successful Payments', color='green')
plt.plot(pending_order_counts['order_date'], pending_order_counts['pending_count'], label='Pending Payments', color='orange')
plt.title('Order Counts by Payment Status Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
